```python
import numpy as np
import phaseportrait
alpha = 0.2
eps   = 0.04
beta  = 1.09
gamma = 1
delta = 0.07

def F(v,w, t=0):
    dv = v * (1 - v) * (v - alpha) - w
    dw = eps * (beta * v - gamma * w - delta)
    return dv,dw

pp4 = phaseportrait.PhasePortrait2D(
    func = F,
    [0,1],
    resolution = 20   # vector grid density
)
pp4.plot()
```
